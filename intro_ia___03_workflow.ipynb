```python
from openai import OpenAI
from google.colab import userdata
api_key = userdata.get('OAIK_JCMV')
client = OpenAI(api_key=api_key)
```

```python
def llm_respuesta(pregunta, condicion):

    # Realiza una consulta a la API de OpenAI.

    response = client.chat.completions.create(
        model="gpt-4.1",
        messages=[
            {"role": "user", "content": pregunta},
            {"role": "system", "content": condicion},
            ]
    )

    generated = response.choices[0].message.content

    return generated
```

```python
import pandas as pd

df = pd.DataFrame([
    {"tema": "Machine Learning",     "prev": "Introducción al análisis de datos", "next": "Técnicas de regresión"},
    {"tema": "Redes Neuronales",     "prev": "Técnicas de regresión",             "next": "Modelos convolucionales"},
    {"tema": "Modelos convolucionales", "prev": "Redes Neuronales",               "next": "Regularización"},
])
```

```python
df
```

```python
condicion = "Asistes con la coherencia de párrafos"
```

```python
def nivel_intermedio(tema):
    return llm_respuesta(f"Redacta un párrafo introductorio sobre {tema}.", condicion)

def nivel_superior(prev, actual, nxt):

    prompt = f"""
    Ajusta este párrafo para asegurar coherencia con el anterior y el siguiente.
    Anterior: {prev}
    Actual: {actual}
    Siguiente: {nxt}
    """
    return llm_respuesta(prompt, condicion)

df["borrador"] = df["tema"].apply(nivel_intermedio)

df["final"] = df.apply(
    lambda row: nivel_superior(row["prev"], row["borrador"], row["next"]),
    axis=1
)

df
```
