```python
# ===============================
# Histogramas de coincidencia
# Tipos de manzana
# ===============================

import pandas as pd
import matplotlib.pyplot as plt

# 1. Crear la tabla de datos (valores en escala 15)
data = pd.DataFrame({
    "tipo": ["Red Delicious", "Granny Smith", "Golden Delicious", "Fuji", "Gala"],
    "dulzor": [4, 1, 4, 5, 4],
    "acidez": [1, 5, 3, 1, 2],
    "textura": [2, 5, 3, 5, 4],
    "conservacion": [2, 3, 2, 3, 2]
})

# 2. Funcin para generar histogramas superpuestos
def histograma_parametro(df, parametro):
    plt.figure()
    for tipo in df["tipo"]:
        valores = df[df["tipo"] == tipo][parametro]
        plt.hist(valores, bins=5, alpha=0.6, label=tipo)
    plt.xlabel(parametro.capitalize())
    plt.ylabel("Frecuencia")
    plt.title(f"Histograma comparativo de {parametro}")
    plt.legend()
    plt.show()

# 3. Generar histogramas para todos los parmetros
parametros = ["dulzor", "acidez", "textura", "conservacion"]

for p in parametros:
    histograma_parametro(data, p)
```
