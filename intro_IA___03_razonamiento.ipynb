# Niveles de Razonamiento o Niveles Cognitivos

```          
           Nivel Cognitivo Superior
                    |
                    v
[ Párrafo anterior ] — (coherencia global) — [ Párrafo siguiente ]
                    ^
                    |
          Nivel Cognitivo Intermedio
                    |
                    v
        Generación del párrafo base
```

````python
from openai import OpenAI
from google.colab import userdata
api_key = userdata.get('OAIK_JCMV')
client = OpenAI(api_key=api_key)
````

````python
def llm_respuesta(pregunta, condicion):

    # Realiza una consulta a la API de OpenAI.

    response = client.chat.completions.create(
        model="gpt-4.1",
        messages=[
            {"role": "user", "content": pregunta},
            {"role": "system", "content": condicion},
            ]
    )

    generated = response.choices[0].message.content

    return generated
````

````python
def generar_parrafo_nivel_intermedio(tema):

    # Genera un párrafo competente sobre el tema. Este es el nivel cognitivo 'intermedio'

    prompt = f"""
    Escribe un párrafo claro, técnico pero comprensible, sobre el siguiente tema:
    '{tema}'. Evita referencias externas y concéntrate en explicar el concepto.
    """
    condicion = f"Eres un asistente que elabora párrafos sobre el {tema}"

    return llm_respuesta(prompt, condicion)


def ajustar_coherencia_nivel_superior(parrafo_anterior, parrafo_actual, parrafo_siguiente, tema):

    # Ajusta el párrafo actual para que sea coherente con el anterior y el siguiente. Este es el nivel cognitivo 'superior'.

    condicion = f"Eres un asistente que elabora párrafos sobre el {tema}"

    prompt = f"""
    Tienes tres párrafos de un manual técnico:

    1) Párrafo anterior:
    {parrafo_anterior}

    2) Párrafo actual (borrador):
    {parrafo_actual}

    3) Párrafo siguiente:
    {parrafo_siguiente}

    Reescribe el párrafo actual para asegurar:
    - Transición lógica desde el párrafo anterior.
    - Coherencia temática y terminológica con el siguiente.
    - Fluidez y claridad.

    Entrega solo el párrafo reescrito.
    """

    return llm_respuesta(prompt, condicion)
````

````python
tema = "Principios básicos de protección del comportamiento climático en el poblado en un valle."

parrafo_prev = "El párrafo explica sobre las condiciones económicas del poblado y sus impactos en los riesgos ambientales"
parrafo_next = "El párrafo explica sobre las consecuencias en la calidad de vida de la población."

# Nivel intermedio
borrador = generar_parrafo_nivel_intermedio(tema)
print(borrador)

# Nivel superior
parrafo_final = ajustar_coherencia_nivel_superior(parrafo_prev, borrador, parrafo_next, tema)

print(parrafo_final)
````

La protección del comportamiento climático en un poblado asentado en un valle se asienta en principios básicos orientados a mitigar los efectos adversos y potenciar los beneficios del microclima local. Debido a la topografía, los valles suelen presentar riesgos particulares como acumulación de aire frío, formación de nieblas y vulnerabilidad ante inundaciones. Para protegerse de estos factores, es fundamental considerar la orientación de las edificaciones, permitiendo la máxima captación de luz solar y ventilación natural, así como la localización estratégica de zonas verdes y cuerpos de agua para regular la temperatura y la humedad. Además, la planificación urbana debe contemplar el diseño de infraestructuras para gestionar el escurrimiento de aguas pluviales y evitar la erosión del suelo. Estos principios contribuyen a crear entornos resilientes y saludables, adaptados a las condiciones particulares de los valles.


En este contexto, resulta esencial abordar cómo el entorno físico y la planificación urbana pueden influir en la reducción de los riesgos ambientales previamente mencionados. En los poblados asentados en valles, la protección frente a los desafíos climáticos específicos—como la acumulación de aire frío, la formación de nieblas o la vulnerabilidad a inundaciones—requiere la aplicación de principios básicos orientados a mitigar efectos negativos y aprovechar los beneficios del microclima local. Estrategias como la adecuada orientación de las edificaciones para optimizar la captación de luz solar y la ventilación natural, la ubicación estratégica de áreas verdes y cuerpos de agua, así como la implementación de infraestructuras para la gestión del agua pluvial y la prevención de la erosión del suelo, son fundamentales. Estas medidas no sólo fortalecen la resiliencia del entorno, sino que también sientan las bases para mejorar la calidad de vida de la población al adaptar el espacio urbano a las condiciones particulares del valle.
